```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras import utils
from sklearn import datasets
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
iris = datasets.load_iris()
name = iris.target_names
x = iris.data
y = iris.target.reshape(-1, 1) 
encoder = OneHotEncoder(sparse=False)
y = encoder.fit_transform(y)
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.20)
```

```python
#Создаём нейронную сеть
model = keras.Sequential([
                     keras.layers.Dense(10, input_shape=(4,), activation='relu'), 
                     keras.layers.Dense(10, activation='relu'), 
                     keras.layers.Dense(3, activation='softmax'), 
])
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(), loss='categorical_crossentropy',metrics= ['accuracy'])
```

```python
print(model.summary())
```

```python
model.fit(train_x, train_y, epochs=250)
```

```python
test_loss, test_acc = model.evaluate(test_x, test_y)
print('Тестовая точность = ', test_acc)
```

```python
prediction = model.predict(train_x)
number = 69
print('Предсказываю = ',name[np.argmax(prediction[number])], '\n','Ответ = ',name[np.argmax(train_y[number])],sep='')
```
